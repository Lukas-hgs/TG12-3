# Position des Sensors im Mülleimer

## Oben

### Vorteile
- Einfach zugänglich
  - Zum Tausch der Batterie
  - Säuberung bei Verschmutzung
- Leichtere Installation
- Bessere Überwachung des Müllstands (zentral)

### Nachteile
- Diebstahl
  - Leicht zu sehen
  - Ungeschützt
- Beschädigung
  - Wenn z. B. ein Pizzakarton reingeschoben wird und am Sensor hängen bleibt
- Verkleinert die Öffnung

#### Sommer
- Wenn den ganzen Tag die Sonne auf den Mülleimer scheint, erhitzt er sich oben stark  
  → mögliche Beschädigung des Sensors oder der Halterung

---

## Unten  
*(unter dem Boden, direkt neben dem Übergang vom Mülleimer zum Hohlraum)*

### Vorteile
- Diebstahlschutz
- Geringere Wahrscheinlichkeit für Beschädigungen  
  (außer eventuell beim Leeren)
- Vor Sonne geschützt

### Nachteile
- Schlechter zugänglich
- Schwerer zu integrieren
- Eventuell schlechtere Überwachung des Müllstands
  - da nicht zentral
- Müll könnte sich ungünstig direkt unter dem Sensor stapeln
- Schlägt schneller Alarm, da näher am Müll
- Wird schneller schmutzig
