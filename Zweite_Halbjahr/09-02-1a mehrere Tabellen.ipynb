#### Aufgabe 1: INNER JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
INNER JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung
Noah fehlt, weil seine projekt_id den Wert NULL hat. Die Wetterstation fehlt, weil kein Schüler diesem Projekt zugeordnet ist.

| Verbindungsart | Zweck |
| --- | --- |
| INNER JOIN | Zeigt nur Zeilen, bei denen der Wert aus schueler.projekt_id zu einem Wert aus projekt.projekt_id passt. |

#### Aufgabe 2: LEFT JOIN


```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
LEFT JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung
Noah ist da obwohl er keinen Titel oder fachbereich hat in den bereichen sthet einfach mull  

| Verbindungsart | Zweck |
| --- | --- |
| Left Join | Zeigt alle Zeilen der Linken Tabelle aund ergänzt nicht passende daten aus der rechten Tabelle mit Null|


#### Aufgabe 3: RIGHT JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
RIGHT JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung

Noah wird nicht mehr angezeigt dafür aber die Wettersation obwohl sie keinen Schüler hat
| Verbindungsart | Zweck |
| --- | --- |
| Right Join | Zeigt alle Zeilen der Rechten Tabelle aund ergänzt nicht passende daten aus der Linken Tabelle mit Null |



#### Aufgabe 4: FULL OUTER JOIN


```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
FULL OUTER JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung

Alles Wird Angezeigt auch Noah und die Wetterstation 
| Verbindungsart | Zweck |
| --- | --- |
| Full OUTER Join | Zeigt alle Zeilen aus beiden tabellen an und ergänzt nicht passende zeilen mit Null |

#### Aufgabe 5: Verbindung mit WHERE ohne JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler, projekt
WHERE schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung


| Verbindungsart | Zweck |
| --- | --- |
|WHERE ohne Join| Verbindet zwei Tabellen über eine Bedingung im der WHERE-Klausel. Entspricht hier einem Inner JOIN |


### Ergebnissicherung
| Verbindungsart | Was wird angezeigt? | Zweck |
| --- | --- | --- |
| INNER JOIN | Nur Daten bei dennen alle Infos vergeben sind |  Zeigt nur Zeilen, bei denen der Wert aus schueler.projekt_id zu einem Wert aus projekt.projekt_id passt. |
| LEFT JOIN | Alle sachen die Links stehen und was es auf der rechten seite nicht gibt wird mit Null aufgefüllt | Zeigt alle Zeilen der Linken Tabelle aund ergänzt nicht passende daten aus der rechten Tabelle mit Null |
| RIGHT JOIN | Alle sachen die Rechts stehen und was es auf der Linken seite nicht gibt wird mit Null aufgefüllt | Zeigt alle Zeilen der Rechten Tabelle aund ergänzt nicht passende daten aus der Linken Tabelle mit Null |
| FULL OUTER JOIN | Zeigt alles an und fehlende Infos werden mit null aufgefüllt |  Zeigt alle Zeilen aus beiden tabellen an und ergänzt nicht passende zeilen mit Null |
| WHERE ohne JOIN | Gleiche wie bei Inner Join aber man muss angeben von wo der die daten nehmen soll | Verbindet zwei Tabellen über eine Bedingung im der WHERE-Klausel. Entspricht hier einem Inner JOIN |
