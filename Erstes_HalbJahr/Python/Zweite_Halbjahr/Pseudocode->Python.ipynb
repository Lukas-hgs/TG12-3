```python
def findelagerplatz(self, pArtikelname):
    idxGefundenLp:int
    idx:int
    name:str
    frei: bool
    gefunden: bool

    idxGefundenLp = -1
    gefunden = False
    idx = 0

    for idx in range (0, self.aAnz,1):
        name = self.derLagerplatz[idx].gibArtikelname()
        if name == pArtikelname:
            gefunden == True
            idxGefundenLp = idx
            self.dieGUI.melden(f"lagerplatz {idxGefundenLp} gefunden")
    
    if gefunden == True:
        for idx in range(0, self.aAnz,1):
            frei = self.derLagerplatz[idx].istFrei()
            if frei == True:
                gefunden == True
                idxGefundenLp = idx
                self.dieGUI.melden("Ware wird eingelagert")
    
    if gefunden == False:
        self.dieGUI.melden("Lager voll, einlagern nicht m√∂glich")
    
    return idxGefundenLp






```
