# Freierfall ohne Luftwiederstand

```python
import math
import matplotlib.pyplot as plt
import numpy as np

t=0
g=9.81
m=80
a=0
v=0
s=0
dt=0.1

tr = []
vr = [] 
sr = []
sg = []

for i in range(0,5):
    t=t+dt
    a=g
    v=v+a*dt
    s=s+v*dt
    s2=0.5*g*t**2

    sg.append(s2)
    tr.append(t)
    vr.append(v)
    sr.append(s)

# Erster Graph: s(t)
plt.plot(tr, sr, color='purple', label='s(t)')
# Zweiter Graph: sg(t)
plt.plot(tr, sg, color='red', label='sg(t)')
plt.xlabel('t(s)')
plt.ylabel('s(m)')
plt.title('Freierfall ohne Luftwiderstand')
plt.grid(True)
plt.legend()
plt.show()
```
