```python
import math

g=9.81
h=0

for i in range(0,10):
    h += 1
    v = math.sqrt(2*g*h)
    print("h:",h,"Meter ""          v:",v,"m/s")




```

```python
import numpy as np


x = np.arange(10,)  # von -10 bis 10 in 0,1er Schritten

print(x)
y = x**2
print(y)
```

```python
from matplotlib import pyplot as plt

plt.plot(x, y, color='purple')  # Farbe der Linie auf Rot setzen
plt.xlabel('x')
plt.ylabel('y')
plt.title('Freierfall: v = √2·g·h')
plt.grid(True)
plt.show()
```
