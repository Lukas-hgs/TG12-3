Aufgabe Verwende die Formeln für den freien Fall aus z.B. 20m Höhe  
• $y(t) = 20 - \frac{1}{2}t²$  
• $v(t) = − gt$  
um das t-y- und das t-v-Diagramm zu erzeugen. Stelle die Diagramme in einem Fenster übereinander da. Wählen
eine sinnvollen Bereich für das Zeit-Array!


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
h = 20  # Anfangshöhe in Metern
g = 9.81  # Erdbeschleunigung in m/s²
v0 = 5 # Anfangsgeschwindigkeit in m/s

# sinnvolles Zeit-Array: von 0 bis zum Auftreffen (y=0)
t_max = np.sqrt(2*h/g)
t = np.linspace(0, t_max, 100)

# Formeln
y = h - 0.5 * g * t**2 + v0 * t
v = -g * t + v0

fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(t, y, label='y(t) = 20 - 0.5gt²', color='green')
ax1.set_ylabel('Höhe y [m]')
ax1.set_xlabel('Zeit t [s]')
ax1.grid(True)
ax1.legend()

ax2.plot(t, v, label='v(t) = - gt', color='blue')
ax2.set_ylabel('Geschwindigkeit v [m/s]')
ax2.set_xlabel('Zeit t [s]')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
