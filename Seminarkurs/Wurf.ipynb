```python
import math
import matplotlib.pyplot as plt
import numpy as np

g = 9.81  # Acceleration due to gravity (m/s^2)
v = 30.0  # Initial velocity (m/s)
t = 0
dt = 0.01  # Time step (s)

winkel = 60

vx = v * math.cos(math.radians(winkel))
vy = v * math.sin(math.radians(winkel))

xR = []
yR = []

for i in range(0,5000):
    t = t + dt
    x = vx * t
    y = vy * t - 0.5 * g * t**2

   
    xR.append(x)
    yR.append(y)
    try:
        if y < 0:
            print(vx)
            print(vy)
            break
    except:
        pass


  



plt.plot(xR, yR)   
plt.title("Schiefer Wurf")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
#plt.ylim(bottom=0)
plt.grid()
plt.show()


```
